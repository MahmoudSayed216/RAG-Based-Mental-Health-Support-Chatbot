```python
from retriever import Retriever
from dotenv import load_dotenv
import os

load_dotenv()
```

```python
ret = Retriever(
    embedding_model=os.getenv('EMBEDDING_MODEL'),
    reranking_model=os.getenv('RERANKING_MODEL'),
    device=os.getenv('DEVICE'),
    vector_db_args={
            "qdrant_client_path": "./qdrant_db",
            "collection_name": os.getenv("EMBEDDINGS_COLLECTION_NAME")
        },
    url=os.getenv('QDRANT_CLUSTER_ENDPOINT'),
    api_key=os.getenv('QDRANT_API_KEY'),
)
```

```python
query = "I've been feeling completely disconnected from everyone around me lately. I lost my job two months ago and since then I just stay in bed all day, I can't find the energy to do anything. I don't know who to talk to or where to even start."

ret.retrieve(query, max_context=2, max_responses=8)
```
